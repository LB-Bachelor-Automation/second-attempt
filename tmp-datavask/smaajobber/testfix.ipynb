```python
from pandas import read_csv, DataFrame, Timedelta, Timestamp, to_datetime
from tkinter import filedialog, Tk

def get_filepath(
    title: str = "Select Raw Data File to Read",
    filetypes: list[dict[str, str]] = [("Raw Data File", ".csv")],
) -> str:
    try:
        #root = Tk()
        #root.withdraw()
        return filedialog.askopenfilename(title=title, filetypes=filetypes)
    except Exception as e:
        print(e)

def set_filepath(
    title: str = "Select Raw Data File to Read",
    filetypes: list[dict[str, str]] = [("Raw Data File", ".csv")],
) -> str:
    try:
        #root = Tk()
        #root.withdraw()
        return filedialog.asksaveasfilename(title=title, filetypes=filetypes)
    except Exception as e:
        print(e)
    

def load_svg_file(filepath:str, sep:str = ","):
    try:
        svg = read_csv(filepath, sep=sep).drop_duplicates()
        reference_time:Timestamp = to_datetime(svg['timestamp'].iloc[0])
        svg["dt"] = [Timedelta(time - reference_time).total_seconds() for time in to_datetime(svg['timestamp'])]
        return svg
    
    except Exception as e:
        print(e)
```

```python
svg = load_svg_file(get_filepath(),";")
```

```python
svg.to_csv(set_filepath())
```
